## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

import ast

# Fill missing values
titles_df['age_certification'] = titles_df['age_certification'].fillna('Unknown')
titles_df['seasons'] = titles_df['seasons'].fillna(0).astype(int)

# Function to safely parse lists from strings
def parse_list(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    elif isinstance(x, list):
        return x
    else:
        return []

# Apply safe parsing
titles_df['genres'] = titles_df['genres'].apply(parse_list)
titles_df['production_countries'] = titles_df['production_countries'].apply(parse_list)

# Export cleaned dataset for visualization
titles_df.to_csv("cleaned_titles_data.csv", index=False)

````

### What all manipulations have you done and insights you found?

1. Handled Missing Values

Filled missing values in age_certification with "Unknown" because many older titles did not have certifications.

Filled missing values in seasons with 0 for movies or shows without season data.

Converted seasons to integer type for consistency.

2. Standardized List Columns

Converted genres and production_countries from string representations (e.g., "['comedy', 'drama']") to actual Python lists.

Used a safe parsing method (ast.literal_eval) and ensured that any malformed or empty entries became empty lists.

Checked that the final genres and production_countries columns are all lists, ready for analysis.

3. Data Type Corrections

Ensured numeric columns like seasons are integers.

Cleaned categorical variables to make them uniform and analysis-ready.

4. Insights from Data Cleaning

Many titles lack age certifications (especially older movies).

Most content is US-produced, but there are some titles from other countries.

genres are multi-valued for most titles, so we need to use exploded views for analysis.

After cleaning, the dataset is ready for visualization, filtering, and aggregation without errors or warnings.
